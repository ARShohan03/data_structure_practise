```python
def sorting(arr):
    sorting=True
    for i in range(len(arr)-1):
                   if arr[i]>arr[i+1]:
                           sorting = False
                           return sorting
                   
    return sorting 

                           
    
```

```python
arr=[1,2,3,4,5]
print(sorting(arr))
```

```python
import time
def monkey_sorting(arr):
    count=0
    while not sorting(arr):
        time.sleep(.001)
        random.shuffle(arr)
        count+=1
        print(f"{count}---->{arr}")
    return arr

```
