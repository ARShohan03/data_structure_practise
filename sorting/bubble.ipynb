```python
def bubbleSorting(arr):
    for i in range(len(arr)-1):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1]=arr[j+1], arr[j]
    return arr
```

```python
a=[39, 54, 656, 34, 22, 1, 44, 3]
print(bubbleSorting(a))
```
