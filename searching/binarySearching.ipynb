```python
def binarySearching(arr, item):
    low=0
    high=len(arr)-1
    print(low, high)
    def find(arr, low, high, item):
            print("skds")
            if low<=high:
                mid=(low+high)//2
                # print(mid) 
                # print(arr[mid])    
                if item == arr[mid]:
                    return mid
                elif item < arr[mid]:
                   return find(arr, low, mid-1, item)
                else:
                    return find(arr, mid+1, high, item)
    result =find(arr, low, high, item)
    if result is None:
         return "not found"
    else:
        return f"found at index {result}"

    
```

```python
arr=[10, 20, 40, 66, 70, 88, 96, 100, 120, 140 ]
```

```python
print(binarySearching(arr, 96))
```
