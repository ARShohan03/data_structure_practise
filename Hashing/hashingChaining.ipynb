```python
class node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next=None
```

```python
class ll:
    def __init__(self):
        self.head=None
    
    def add(self, key, value):
        new_node = node(key, value)

        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while temp.next is not None:
                temp=temp.next
            temp.next=new_node

    def deleteHead(self):
        if self.head==None:
            return "noting to delete"
        else:
            self.head=self.head.next

    def remove(self, key):
        if self.head is None:
            return "empty"
        if self.head.key==key:
            self.deleteHead()
            return 
       
        temp=self.head
        while temp.next is not None:
            if temp.next.key==key:
                temp.next=temp.next.next
                return 
            temp=temp.next
        return "key is not found"
    
    def traverse(self):
        temp=self.head
        while temp is not None:
            print(f"key: {temp.key}, value: {temp.value}")
            temp=temp.next

    def size(self):
        count=0
        temp=self.head
        while temp is not None:
            count+=1
            temp=temp.next
        return count
    
    def search(self, key):
        temp=self.head
        pos=0

        while temp is not None:
            if temp.key==key:
                return pos
            
            temp=temp.next

            pos+=1
        return -1
    def get_node_at_index(self, nodeIndex):
        temp=self.head
        count=0

        while temp is not None:
            if count==nodeIndex:
                return temp
            temp=temp.next
            count+=1
        return
```

```python
class dictionary:
    def __init__(self, capacity):
        self.size=0
        self.capacity=capacity
        self.buckets=self.makeArray(capacity)

    def makeArray(self, capacity):
        l=[]

        for i in range(capacity):
            l.append(ll())
        return l
    
    def put(self, key, value):
        bucketIndex=self.hashFunction(key)
        
        nodePos=self.buckets[bucketIndex].search(key)
        if nodePos==-1:
            self.buckets[bucketIndex].add(key, value)
            self.size+=1
            loadFactor=self.size/self.capacity
            print("Load fator=", loadFactor)

            if loadFactor>=2:
                self.rehash()      
        else:
            node=self.buckets[bucketIndex].get_node_at_index(nodePos)
            node.value=value

        
        print("final bucket len =", len(self.buckets))

        return 
    
    # def get(self, key):
    #     bucketIndex=self.hashFunction(key)
    #     for i in range(len(self.buckets)):
    #         nodePos=self.buckets[i].search(key)
    #         node=self.buckets[i].get_node_at_index(nodePos)
    #         print(f"{node.key}, :, {node.value}")
    def get(self, key):
        bucketIndex=self.hashFunction(key)
        nodePos=self.buckets[bucketIndex].search(key)
        
        if nodePos==-1:
            return "node is not found"
        else:
           node = self.buckets[bucketIndex].get_node_at_index(nodePos)
           return node.value


    def hashFunction(self, key):
        return abs(hash(key))%self.capacity
    
    def rehash(self):
        oldBuckets = self.buckets
        print("old bucket len =", len(oldBuckets))

        self.capacity = self.capacity*2
        self.size=0
        newBuckets=self.makeArray(self.capacity)

        self.buckets=newBuckets

        print("new bucket len =", len(newBuckets))


        # print(len(oldBuckets))

        for i in range(len(oldBuckets)):
            for j in range(oldBuckets[i].size()):
                print("start", f"oldBuckets[{i}] :" , oldBuckets[i].traverse(), "end")
                node=oldBuckets[i].get_node_at_index(j)
                self.put(node.key, node.value)

    def __setitem__(self, key, value):
        return self.put(key, value)
    
    def __getitem__(self, key):
        return self.get(key)
    
    def __str__(self):
        for i in range (len(self.buckets)):
            self.buckets[i].traverse()
        return ""
        
    def __len__(self):
        return self.size
    
    def __delitem__(self, key):
        bucketIndex=self.hashFunction(key)
        self.buckets[bucketIndex].remove(key)
        self.size-=1


```
