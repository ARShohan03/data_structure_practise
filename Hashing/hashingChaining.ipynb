```python
class node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next=None
```

```python
class ll:
    def __init__(self):
        self.head=None
    
    def add(self, key, value):
        if self.head is None:
            new_node = node(key, value)
        else:
            temp=self.head
            while temp.next is not None:
                temp=temp.next
            temp.next=new_node

    def deleteHead(self):
        if self.head==None:
            return "noting to delete"
        else:
            self.self.head=self.head.next

    def remove(self, key):
        if self.head is None:
            return "empty"
        if self.head.key==key:
            self.deleteHead()
            return 
       
        temp=self.head
        while temp.next is not None:
            if temp.next.key==key:
                temp.next=temp.next.next
                return 
            temp=temp.next
        return "key is not found"
    
    def traverse(self):
        temp=self.head
        while temp is not None:
            print(f"key: {temp.key}, value: {temp.value}")
            temp=temp.next

    def size(self):
        count=0
        temp=self.head
        while temp is not None:
            count+=1
            temp=temp.next
        return count
    
    def search(self, key):
        temp=self.head
        pos=0

        while temp is not None:
            if temp.key==key:
                return pos
            
            temp=temp.next

            pos+=1
        return -1
    def get_node_at_index(self, nodeIndex):
        temp=self.head
        count=0

        while temp is not None:
            if count==nodeIndex:
                return temp
            temp=temp.next
            count+=1
        return
```

```python
class dictionary:
    def __init__(self, capacity):
        self.size=0
        self.capacity=capacity
        self.buckets=self.makeArray(capacity)

    def makeArray(self, capacity):
        l=[]

        for i in range(capacity):
            l.append(ll())
        return l
    
    def put(self, key, value):
        bucketIndex=self.hashFunction(key)
        
        nodePos=self.buckets[bucketIndex].search(key)
        if nodePos==-1:
            self.buckets[bucketIndex].add(key, value)
        else:
            node=self.buckets[bucketIndex].get_node_at_index(self.buckets[bucketIndex].search(key))
            node.value=value
        loadFactor=self.size/self.capacity

        if loadFactor>=2:
            self.rehash()  
        
        return 
    
    def get(self, key):
        bucketIndex=self.hashFunction(key)
        for i in range(len(self.buckets)):
            nodePos=self.buckets[i].search(key)
            node=self.buckets[i].get_node_at_index(nodePos)
            print(f"{node.key}, :, {node.value}")



    def hashFunction(self, key):
        return abs(hash(key))%self.capacity
    
    def rehash(self):
        oldBuckets = self.buckets
        newBuckets = self.buckets*2
        self.buckets=newBuckets

        for i in range(len(oldBuckets)):
            for j in range (oldBuckets[i].size()):
                node=oldBuckets[i].get_node_at_index(j)
                self.put(node.key, node.value)

    def __set__(self, key, value):
        return self.put(key, value)
    
    def __get__(self, key):
        return self.get(key)
    
    def __str__(self):
        for i in range (len(self.buckets)):
            self.buckets[i].traverse()
            return ""
        
    def __len__(self):
        return self.size
    
    def __del__(self, key):
        for i in range(len(self.buckets)):
            self.buckets[i].remove(key)



```
