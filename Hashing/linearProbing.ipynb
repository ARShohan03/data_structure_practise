```python
class dictionary:
    def __init__(self, size):
        self.size=size
        self.slot=[None]*self.size
        self.data=[None]*self.size
    def put(self, key, value):
        hashValue=self.hash_func(key)

    # if the slot is empty, insert directly
        if self.slot[hashValue] is None:
            self.slot[hashValue]=key
            self.data[hashValue]=value
        # if the slot isn't empty, check if the key is same
        else:
            if self.slot[hashValue]==key:
                self.data[hashValue]=value
            #linear probing, going forward to find the next slot
            else: 
                newHashValue=self.rehash(hashValue)
            # going forward continuously untill a free slot is found or the same key is being found in the slot 
                while self.slot[newHashValue]!=None and self.slot[newHashValue]!=key:
                    newHashValue=self.rehash(newHashValue)
                
                if self.slot[newHashValue]==None:
                    self.slot[newHashValue]=key
                    self.data[newHashValue]=value
                if self.slot[newHashValue]==key:
                    self.data[newHashValue]=value
                

                
    def hash_func(self, key):
        return abs(hash(key))%self.size
    def rehash(self, oldHashvalue):
        return (oldHashvalue+1)%self.size
    
    def get(self, key):
        startPos=self.hash_func(key)
        curr=startPos

        while self.slot[curr] is not None:
            print("aaaaa")
            if self.slot[curr]==key:
                return self.data[curr]
            
            curr=self.rehash(curr)
            print("rehash done")
            if curr==startPos:
                print("bbbbbb")
                print(startPos)
                return "data is not there"

        return "None found no data"
    def __setitem__(self, key, value):
        return self.put(key,value)
    def __getitem__(self, key):
        return self.get(key)
    
    def __str__(self):
        for  i in range(len(self.slot)):
            if self.slot[i] is not None:
                print(self.slot[i], ":", self.data[i], end=" ")
        return ""



```
